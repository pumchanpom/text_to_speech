```python
# Text to Speech Function and Loop
# gTTS Library Installation
!pip install gtts
!pip install IPython
import gtts
from IPython.display import Audio
```

```python
# Texts
unit2 = """1 adventure
2 favorite
3 common
4 connect
5 grow-up
6 details
7 opportunity
8 seasonal
9 share
10 special """

unit3 = """ - situated
- down to earth
- vendors
- stalls
- genuine
- attractions
- countless
- floating
- condiments
- narrow
- booth
- located
- seasoning
- practical
- pure
- natural
- numerous
- drifting
- cramped
- places of interests
- sellers """

unit4 = """
- acquaintance
- reunion
- catch-up
- literature
- humid
- skyscraper
- hybrid
- bound to
- undoubtedly
- spontaneous
"""
```

```python
# Functions
def make_audio(text_to_sound, filename):
    gTTSobject = gtts.gTTS(text = text_to_sound,
                       tld ="com",
                       lang ="en",
                       slow = False)
    gTTSobject.save(filename)
units = [(unit2, "unit2.mp3"), (unit3, "unit3.mp3"), (unit4, "unit4.mp3")]
```

```python
# Play audio
for unit, filetitle in units:
    make_audio(unit, filetitle)
    audio = Audio(filename=filetitle, autoplay=False)
    display(audio)
```
